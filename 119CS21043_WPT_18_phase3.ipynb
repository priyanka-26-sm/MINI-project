```python
import pandas as pd
df=pd.read_excel('E:/chrome downloads/onlineretail/online_retail_II.xlsx')
df.head()
```

## 1. Univariate Analysis

```python
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame with the dataset
# Assuming 'df' has a column named 'Quantity'

# Summary statistics
summary_stats = df['Quantity'].describe()

# Histogram for Quantity
plt.figure(figsize=(8, 6))
plt.hist(df['Quantity'], bins=20, edgecolor='k')
plt.title('Histogram of Quantity')
plt.xlabel('Quantity')
plt.ylabel('Frequency')
plt.show()

# Box plot for Quantity
plt.figure(figsize=(8, 6))
plt.boxplot(df['Quantity'], vert=False)
plt.title('Box Plot of Quantity')
plt.xlabel('Quantity')
plt.show()

print(summary_stats)
```

## 2. Bivariate Analysis

```python
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame with the dataset
# Assuming 'df' has columns named 'Quantity' and 'Price'

# Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(df['Quantity'], df['Price'], alpha=0.5)
plt.title('Scatter Plot of Quantity vs. Price')
plt.xlabel('Quantity')
plt.ylabel('Price')
plt.show()

# Correlation coefficient
correlation = df['Quantity'].corr(df['Price'])
print(f"Correlation coefficient: {correlation}")
```

## 3. Multivariate Analysis

```python
import seaborn as sns

# Assuming 'df' is your DataFrame with the dataset

# Scatter plot matrix
sns.pairplot(df[['Quantity', 'Price', 'Customer ID']], diag_kind='kde')
plt.show()
```

```python
# Correlation heatmap
corr_matrix = df[['Quantity', 'Price', 'Customer ID']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

## 4.    Visualize above with Bar, Box, FacetGrid, Heatmap, Crosstab,Line plot Submission

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame with the dataset

# Bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x='Country', y='Quantity', data=df)
plt.title('Bar Plot of Quantity by Country')
plt.xlabel('Country')
plt.ylabel('Quantity')
plt.show()
```

```python
# Box plot
plt.figure(figsize=(8, 6))
sns.boxplot(x='Country', y='Quantity', data=df)
plt.title('Box Plot of Quantity by Country')
plt.xlabel('Country')
plt.ylabel('Quantity')
plt.show()
```

```python
# Crosstab
cross_tab = pd.crosstab(df['Country'], df['StockCode'])
plt.figure(figsize=(10, 8))
sns.heatmap(cross_tab, cmap='YlGnBu', annot=True, fmt='d')
plt.title('Crosstab Heatmap')
plt.xlabel('StockCode')
plt.ylabel('Country')
plt.show()
```

```python
# Heatmap
corr_matrix = df[['Quantity', 'Price', 'Customer ID']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python

# Line plot (assuming 'InvoiceDate' is converted to datetime format)
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df.set_index('InvoiceDate', inplace=True)
df.resample('M').sum()['Quantity'].plot(figsize=(10, 6))
plt.title('Monthly Total Quantity')
plt.xlabel('Date')
plt.ylabel('Total Quantity')
plt.show()
```

```python
# FacetGrid
g = sns.FacetGrid(df, col='Country', margin_titles=True)
g.map(plt.hist, 'Quantity', color='steelblue', bins=20)
plt.show()
```
